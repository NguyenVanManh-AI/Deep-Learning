```python
import numpy as np
import pandas as pd

# Đọc dữ liệu từ file CSV
df = pd.read_csv('data_logistic.csv')

# Chuyển dữ liệu thành mảng numpy
data = df.to_numpy()

# Tách dữ liệu thành input và output
X = data[:, :-1]  # Lấy tất cả các cột trừ cột cuối cùng là output
y = data[:, -1]   # Lấy cột cuối cùng là output

# Hàm sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Đạo hàm của sigmoid
def sigmoid_derivative(x):
    return x * (1 - x)

# Khởi tạo trọng số ngẫu nhiên
np.random.seed(1)
weights = 2 * np.random.random((2, 1)) - 1

# Training
for epoch in range(10000):
    # Feedforward
    input_layer = X
    outputs = sigmoid(np.dot(input_layer, weights))
    
    # Tính toán lỗi
    error = y.reshape(-1, 1) - outputs
    
    # Tính điều chỉnh cần thiết cho trọng số
    adjustments = error * sigmoid_derivative(outputs)
    
    # Cập nhật trọng số
    weights += np.dot(input_layer.T, adjustments)

# In ra kết quả
print("Kết quả sau khi huấn luyện:")
print(outputs)
```
